```python
df['model'] = df['model'].fillna('Unknown Product')
```

```python
df= df.fillna(0)
```

```python
df.isnull().sum()
```

```python
top_brands = df['make'].value_counts().head(10)

plt.figure()
top_brands.plot(kind='bar')
plt.title('Top 10 Car Brands by Sales Count')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\nInsight: Hyundai and Infiniti have the highest sales count.")
```

```python
plt.figure()
plt.scatter(df['odometer'], df['mmr'])
plt.title('Odometer vs mmr')
plt.xlabel('Odometer')
plt.ylabel('mmr')
plt.tight_layout()
plt.show()

print("Insight: As mileage increases, mmr decreases (negative relationship).")
```

```python
transmission_counts = df['transmission'].value_counts()

plt.figure()
transmission_counts.plot(kind='bar')
plt.title('Transmission Type Distribution')
plt.xlabel('Transmission')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print("Insight: Automatic cars dominate the dataset.")
```
